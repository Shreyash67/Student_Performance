```python
import pandas as pd
```

```python
df = pd.read_csv("H:\Student_Performance\data\StudentsPerformance.csv")
```

```python
df.head()
```

```python
df.drop(columns = {'race/ethnicity','parental level of education','lunch','test preparation course','gender'},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"math score":"math_score","reading score":"reading_score","writing score":"writing_score"},inplace=True)
```

```python
df['Average'] = df[['math_score','reading_score','writing_score']].mean(axis=1)
```

```python
df.head()
```

```python
df['Result'] = ''
```

```python
df.head()
```

```python
for i in range(len(df)):
    if df.loc[i,"Average"]>=50:
        df.loc[i,"Result"] = "Pass"
    else:
        df.loc[i,"Result"] = "Fail"
```

```python
df.head()
```

```python
df.drop(columns={'Average'},axis=1,inplace=True)
```

```python
df.head()
```
